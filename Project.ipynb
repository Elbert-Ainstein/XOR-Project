### Part F

A neuron's decision rule is **w1​x1 ​+ w2​x2 ​+ b > 0**. Geometrically it is a **line** on the **(x1, x2) plane**, and the points on one side fire where the other side are shut down. The weights are the slope-setters while bias slide intercepts around. So a single neuron is, structurally, a *line-drawer*. So theoretically a neuron's limit, instead of being within its cleverness of picking weights, it deals more of the **shape** of the data being processed.

### Part A

The XOR classes cannot be separated by any given line, unless we **transform** the input space first using a new layer. Then we can draw its line in this transformed space. The hidden layer mapped the four XOR points into new coordinates where the two 1-points ended up stacked at (1,1) and the 0-points landed elsewhere. This allows the neuron to actually work.

### Part B

Each hidden neuron is a feature detector, with N1 being an OR gate and N2 being a NAND gate. Neither of them forms the XOR structure, but it forms a part of it, and the output neuron can work with these intermediate features that are more defined and friendly.

### Part C

Linear/Logistic regression are essentially neurons, as when you stack them the output of one regression can be the input of another one. However, such stacking can allow us to evolve from fitting a line to fitting more complex shapes. 

### Part D

Each layer passes a yes/no answer forward from a weighted question about its inputs, and then layer by layer the questions/answers gets propelled forward until the output neuron puts everything together in a verdict. A decision is a chain of small linear judgments separated by nonlinear gates, where one neuron does not do much but their combinations create promising results

### Part E
