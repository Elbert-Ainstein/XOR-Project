### Part B
#### Part i
We need four conditions fulfilled at the same time.
- (0, 0): z = b, must be ≤ 0.
- (0, 1): z = w2 + b, must be ≤ 0.
- (1, 0): z = w1 + b, must be ≤ 0.
- (1, 1): z = w1 + w2 + b, must be > 0.
A clean solution is w1 = w2 = 1, and b = -1.5

#### Part ii
The reason why the bias is negative is because it acts as the reluctance to fire. b being -1.5 means that one input is not enough to fire as at most it will be z = 1 + b = -0.5 -> no firing.
